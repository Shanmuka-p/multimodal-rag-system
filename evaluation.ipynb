# Multimodal RAG System Evaluation

This notebook is designed to test and evaluate the functionality and performance of the Multimodal RAG API. It provides a structured way to send various types of queries and inspect the generated answers and their sources.

```python
import requests
import json
from PIL import Image
import io
import base64

def print_json(data):
    """Helper function to pretty-print JSON responses."""
    print(json.dumps(data, indent=2))
```

## 1. Configuration

First, let's define the location of our running API. The default is `http://localhost:8000`.

```python
API_URL = "http://localhost:8000/query"
```

## 2. Helper Function

This function sends a query to the API and prints the response. It also handles potential connection errors.

```python
def query_rag_system(query: str, top_k: int = 5):
    """Sends a query to the RAG API and prints the results."""
    payload = {
        "query": query,
        "top_k": top_k
    }
    headers = {
        "Content-Type": "application/json"
    }
    
    print(f"Sending Query: '{query}'...\n")
    
    try:
        response = requests.post(API_URL, headers=headers, json=payload)
        response.raise_for_status() # Raises an exception for bad status codes (4xx or 5xx)
        
        result = response.json()
        
        print("--- Generated Answer ---")
        print(result.get("answer"))
        print("\n--- Sources ---")
        print_json(result.get("sources", []))
        
    except requests.exceptions.RequestException as e:
        print(f"Error connecting to the API: {e}")
        print("Please ensure the FastAPI server is running and accessible at the specified URL.")
```

## 3. Evaluation Queries

Now, let's run a few queries to test the system. These queries are designed to check different capabilities, such as extracting simple text, interpreting visuals, and combining information from both.

### Query 1: Text-Based Question

This query should be answerable from the text content in one of the sample documents.

```python
# Replace with a question relevant to your sample documents
text_query = "What is the main topic of the document?"
query_rag_system(text_query)
```

### Query 2: Image-Based Question

This query should require the system to 'look' at an image, chart, or diagram in one of the documents.

```python
# Replace with a question about a visual element in your documents
image_query = "Describe the chart shown on page 3."
query_rag_system(image_query)
```

### Query 3: Combined Information Question

This query tests the system's ability to synthesize an answer from both text and visual information.

```python
# Replace with a complex question requiring data from multiple sources
combined_query = "Based on the text and the graph, what was the trend in revenue?"
query_rag_system(combined_query)
```
